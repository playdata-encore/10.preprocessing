### 조건 기반 필터링

```python
import numpy as np
import pandas as pd
```

```python
college = pd.read_csv('0.dataSet/college.csv')
college
```

```python
# college의 주가 'TX' 텍사스인 데이터에 한해서만
college.columns
```

```python
# 텍사스 주에 있는 대학
c1 = college.STABBR == 'TX'
```

```python
# 대학에 백인 비율이 50% 이상만 검색 
c2 = college.UGDS_WHITE >= 0.5
```

```python
c3 = c1 & c2
c3
```

```python
# row 가체가 boolean으로 적용 따라서 True인 row만 filtering
college[c3]
```

```python
cols = ['CITY', 'STABBR']
college.loc[c3, cols]
```

```python
cr2 = college.UGDS_WHITE >= 0.5
cr3 = college.UGDS_WHITE <= 0.8
cr4 = cr2 + cr3
cr4
```

```python
college[cr4]
```

```python
# 상단과 동일한 로직의 문장 - 숫자 포함 관계 처리 필터링 : between
cr5 = college.UGDS_WHITE.between(0.5, 0.8)
cr5
```

```python
college[cr5]
```

```python
# TX와 MI값이 이상치라 할 경우라 간주, 제거해야 하는 상황이라면 어떻게? ~ 표기의 부정 연산자로 필터링

states = ['TX', 'MI']
cr1 = ~college.STABBR.isin(states)
cr2 = college.UGDS_WHITE.between(0.5, 0.8)
cr = cr1+cr2
college[cr]
```
