```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# test를 위한 정규분포 범위내의 난수 발생
dict(time=np.arange(500), value=np.random.randn(500))
```

```python
np.arange(500).cumsum()
```

```python
dict(time=np.arange(500), value=np.arange(500).cumsum())
```

```python
df = pd.DataFrame(dict(time=np.arange(500), value=np.random.randn(500).cumsum()))
df.head(10)
```

```python
sns.relplot(x='time', y='value' , data=df)
```

```python
sns.relplot(x='time', y='value' , data=df, kind='line')
```

```python
sns.relplot(x='time', y='value' , data=df, kind='scatter')
```

**fmri - 혈류의 변화와 관련된 뇌 활동 측정하는 기술**

1. subject : 사람
2. timeoint : 측정 시점
3. event : 활동 종류
4. region
5. signal : 측정값

```python
# MRI의 측정값중 하나의 센터값을 정리한 시계열 데이터 

fmri = sns.load_dataset('fmri')
fmri
```

```python
fmri.head(20)
```

반복적으로 관측된 데이터를 기반으로 신뢰구간, 추정 구간 도출 가능

```python
sns.relplot(x='timepoint', y='signal', data=fmri, kind='line');
```

```python
sns.relplot(x='timepoint', y='signal', data=fmri, kind='line', ci=None);
```
