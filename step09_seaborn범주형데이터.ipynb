```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
tips
```

```python
# 요일별로 total_bill 표현
# total_bill	tip	sex	smoker	day	time	size

sns.catplot(x='day', y='total_bill', data=tips)
```

```python
sns.catplot(x='day', y='total_bill', data=tips, kind='swarm')
```

```python
sns.catplot(x='day', y='total_bill', data=tips, kind='box')
```

```python
sns.catplot(x='day', y='total_bill', data=tips, kind='violin')
```

```python
sns.catplot(x='day', y='total_bill', data=tips, kind='boxen')
```

```python
sns.catplot(x='day', y='total_bill', data=tips, kind='bar')
```

```python
sns.catplot(x='day', y='total_bill', data=tips, kind='bar', hue="day")
```

```python
sns.catplot(x='day', y='total_bill', data=tips, kind='point')
```

```python
sns.catplot(x='day', y='total_bill', data=tips, kind='swarm', hue='sex')
```

```python
sns.catplot(x='day', y='total_bill', data=tips, kind='violin', hue='sex')
```

```python
sns.catplot(x='day', y='total_bill', data=tips, kind='violin', hue='sex', split=True)
```

```python
titanic = sns.load_dataset('titanic')
titanic
```

```python
# 등급별(class), 성별(sex), 생존(survived) 데이터를 bar chart로 표현

sns.catplot(x='sex', y='survived', hue='class', data=titanic, kind='bar')
```

```python
sns.catplot(x='sex', y='survived', hue='class', data=titanic, kind='bar', ci=None)
```
